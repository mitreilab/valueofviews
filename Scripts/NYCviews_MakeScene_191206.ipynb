# Functions (old -- not using in this script)

```python
########################################################################  
######## FUNCTIONS ######## 
######################################################################## 

# Get a list of the immediate subdirectories in a folder
def get_immediate_subdirs(a_dir):
    return [name for name in os.listdir(a_dir)
            if os.path.isdir(os.path.join(a_dir, name))]
    
    
#Revise the 4th line in OBJ file to include the material name
def reviseObjMaterial(material, filepath, writepath): 
    
    # material is name of the material to match the material definition
    # filepath is the directory of the obj files **include 'r' in front to make raw string!
    # writepath is the directory where to write the revised obj files
    
    for path, subdirs, files in os.walk(filepath):
        for filename in files: 
            if filename.endswith('.obj'): 
                
                # Rewrite the material name
                with open(path+'\\'+filename, 'r') as f: 
                    orig = f.readlines()
      
                    # Rewrite the third line in text file
                    orig[2] = 'o ' + material + '\n'                  
                    
                    # Split lines into list to switch places of Y and Z values
                    new_text = []
                    for line in orig: 
                        line = line.split()
                        
                        if len(line) >= 3: 
                            if line[0] == 'v': 
                                line[2], line[3] = str(-float(line[3])), line[2]
                            
                            if line[0] == 'vn': 
                                if line[3] == '0': 
                                    line[2], line[3] = line[3], line[2]
                                else: 
                                    line[2], line[3] = str(-float(line[3])), line[2]
                        
                        # Transform list of lists to list of strings
                        new_text.append(' '.join(line)+'\n')

                    # Save as a new file 
                    id_name = filename[:-4]
                    with open(writepath+'\\'+ id_name + '_rev.obj', 'w') as outfile: 
                        for pos, line in enumerate(new_text):
                            #Omit line 4 in the original text file
                            if pos != 3: 
                                outfile.write(line)   


#Revise the 4th line in OBJ file to include the material name
def reviseObjMaterialBldgs(filepath, writepath): 

    # Dictionary with the materials for each building part
    bldgPart = {'ceiling': 'GenericCeiling_70', 'contextMeshes': 'OutsideFacade_30', 'floor':'GenericFloor_20', 'glazing': 'Glazing_SinglePane_88', 'walls': 'GenericInteriorWall_50'}

    for path, subdirs, files in os.walk(filepath):
        for filename in files: 
            if filename.endswith('.obj'):
                
                with open(path+'\\'+filename, 'r') as f: 
                    orig = f.readlines()
                    
                    # Determine building part and rewrite the third line in text file
                    for key, value in bldgPart.items(): 
                        if key in filename: 
                            orig[2] = 'o ' + value + '\n'
                
                    # Split lines into list to switch places of Y and Z values
                    new_text = []
                    for line in orig: 
                        line = line.split()
                        
                        if len(line) >= 3: 
                            if line[0] == 'v': 
                                line[2], line[3] = str(-float(line[3])), line[2]
                            
                            if line[0] == 'vn': 
                                if line[3] == '0': 
                                    line[2], line[3] = line[3], line[2]
                                else: 
                                    line[2], line[3] = str(-float(line[3])), line[2]
                        
                        # Transform list of lists to list of strings
                        new_text.append(' '.join(line)+'\n')
                    
                    # Save as a new file 
                    id_name = filename[:-4]
                    with open(writepath+'\\'+ id_name + '_rev.obj', 'w') as outfile: 
                        for pos, line in enumerate(new_text):
                            if pos != 3: 
                                outfile.write(line)   
```
